```python
import pandas as pd


#display all output form a cell not just the last (the options are 'all', 'none', 'last' and 'last_expr'.)

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
PIT = {}
for i in range(2014, 2021):
    PIT[str(i)] = pd.read_excel('../2007-2020-PIT-Estimates-by-CoC.xlsx', sheet_name = str(i))
    
```

```python
#More demographic, gender information collected starting in 2015 and then in 2017

[(year,df.shape) for year, df in PIT.items()]

#rows without CoC Number

[df['CoC Number'].isna().sum() for df in PIT.values()]
```

```python
for year in PIT.keys():
    #stripping year from column names
    
    assert(len(PIT[year].columns) == len([col for col in PIT[year].columns.str.strip(', ' + year)]))
    PIT[year] = PIT[year].rename(columns = {col:col.replace(', '+ year, '') for col in PIT[year].columns })
    
    #dropping rows with no CoC Number
    #these are rows corresponding to the totals and a footer
    
    PIT[year] = PIT[year].dropna(subset = ['CoC Number'])
    
    #adding a year column
    
    PIT[year]['year'] = int(year)
```

```python
[df['CoC Number'].isna().sum() for df in PIT.values()]
```

```python
[[col for col in PIT['2020'].columns if col not in PIT[year].columns] for year in ['2017','2018','2019']]
```

```python
#collated dataframe

data = pd.concat([PIT[year] for year in PIT.keys()])
```

```python
data.shape
data.head()
```

```python
#Adding city and state to data

cities = pd.read_csv('./HUDNum_City_imputed - Sheet1.csv')

cities[cities['City'].isna()]

cities.shape
cities = cities.dropna(subset = ['City'])
cities.shape
cities['City'] = cities['City'].str.lower()
cities['CocState'] = cities['CocState'].str.lower()
cities = cities.rename(columns = {'CocState': 'State'})
```

```python
data = data.merge(cities, how = 'left', on = ['CoC Number', 'CoC Name'])
```
