```python
cor = df1.corr().select_dtypes(np.number)['overall homeless, 2020']
cor
```

```python
df2 = df1.drop(columns=['rank_y', 'rank', 'unnamed: 0'])
corr_matrix = df2.corr()

plt.figure(figsize =(10,10))
plt.subplot(1,2,1)
sns.heatmap(corr_matrix[['overall homeless, 2020']].sort_values(by = 'overall homeless, 2020',ascending = False).head(10),
            annot = True,
            vmin = -1,
            vmax = 1,
            cmap = 'Greens');
plt.subplot(1,2,2)
sns.heatmap(corr_matrix[['overall homeless, 2020']].sort_values(by = 'overall homeless, 2020',ascending = True).head(10),
            annot = True,
            vmin = -1,
            vmax = 1,
            cmap = 'Greens');
plt.tight_layout()
```
