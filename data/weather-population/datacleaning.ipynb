```python
ballotpedia = pd.read_csv('ballotpedia_population_info.csv')
citymayors = pd.read_csv('population_info_citymayors.csv')
weatherdata= pd.read_csv('WeatherData.csv')
wpr = pd.read_csv('worldpopulationreview.csv')
```

```python
def remove_state(name):
    chars = []
    for char in name:
        if char == ',' or char == ';': break
        chars.append(char)
    return ''.join(chars).replace('New York City', 'New York')
        
```

```python
ballotpedia['City'] = [remove_state(city) for city in ballotpedia['City']]
citymayors['City; State'] = [remove_state(city) for city in citymayors['City; State']]
weatherdata['City'] = [remove_state(city) for city in weatherdata['City']]
wpr['name'] = [remove_state(city) for city in wpr['name']]

ballotpedia.rename({'City' : 'city'}, inplace = True, axis = 1,)
citymayors.rename({'City; State': 'city'}, inplace = True, axis = 1,) 
weatherdata.rename({'City':'city'}, inplace = True, axis = 1,)
wpr.rename({'name':'city'}, axis = 1, inplace = True) 
```

```python
df = pd.merge(weatherdata, wpr, on = 'city', how = 'left')
df = pd.merge(df, citymayors, on = 'city', how = 'left')
df = pd.merge(df, ballotpedia, on = 'city', how = 'left')
df.shape
```
