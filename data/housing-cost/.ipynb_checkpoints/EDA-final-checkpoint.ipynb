# Linear Regression model

```python
X = df_2019_w_homless.drop(columns=['overall homeless', 'year','city', 'state'])
y = df_2019_w_homless['overall homeless']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
def lr_coef(year):
    year_str = str(year)
    lr_year_str = LinearRegression()
    lr_year_str.fit(X_train, y_train)

    print(lr_year_str.score(X_train, y_train))
    #print(type(year))
    return pd.DataFrame({
    'Features': X_train.columns,
    'Coef': lr_year_str.coef_
    })
```

```python
lr_coef(2019)
```

## For 2019

```python
lr_coef(2019)
```

```python
X = df_2019_w_homless.drop(columns=['overall homeless', 'year','city', 'state'])
y = df_2019_w_homless['overall homeless']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

lr_2019 = LinearRegression()
lr_2019.fit(X_train, y_train)

print(lr_2019.score(X_train, y_train))

pd.DataFrame({
    'Features': X_train.columns,
    'Coef': lr_2019.coef_
})
```
